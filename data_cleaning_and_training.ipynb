```python
import string
from nltk.corpus import stopwords
filepath = 'wikipedia_documents_data/Alan_Turing.txt'
with open(filepath, 'r') as f:
    data = f.read()
    
# Stopwords removal and lower the text 
data = data.translate(data.maketrans("", "", string.punctuation)).lower()
# text = "The quick BROWN Fox Jumps OVER the lazy dog."

stop_words = set(stopwords.words("english"))
words = data.split()
filtered_words = [word for word in words if word not in stop_words]
text = " ".join(filtered_words) 
print(text)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

words = text.split(" ")
stemmed_words = [stemmer.stem(word) for word in words]
text = ",".join(stemmed_words)
print(text)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = text.split(",")
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
text = " ".join(lemmatized_words)
print(text)
```

```python
from nltk.corpus import wordnet as wn
def get_synonyms(word):
    synonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name().replace('_', ' '))
    return list(synonyms)
```

```python
from tqdm import tqdm

# Step 2: Create a dictionary of possible synonyms for common words in the text
def generate_synonym_dict(text_column):
    synonym_dict = {}
    words = set(text_column)  # Get unique words from the text --- " ".join(text_column).split()
    for word in tqdm(words, desc="Generating synonyms"):
        synonyms = get_synonyms(word)
        if synonyms:  # Only add if synonyms are found
            synonym_dict[word] = synonyms  # Use the first synonym as replacement
    return synonym_dict
```
